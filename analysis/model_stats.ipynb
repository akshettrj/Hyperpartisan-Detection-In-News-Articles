```python
# Load all the data
elmo_epochs = []
bert_subj_epochs = []
elmo_subj_epochs = []
bert_epochs = []
with open("./stats/baseline_epochs.txt", "r") as data_file:
    elmo_epochs = data_file.readlines()
with open("./stats/baseline_plus_epochs.txt", "r") as data_file:
    elmo_subj_epochs = data_file.readlines()
with open("./stats/bert_normal_epochs.txt", "r") as data_file:
    bert_epochs = data_file.readlines()
with open("./stats/bert_subjective_epochs.txt", "r") as data_file:
    bert_subj_epochs = data_file.readlines()
```

```python
# Gather data
elmo_train_losses = []
elmo_train_accuracies = []
elmo_val_losses = []
elmo_val_accuracies = []

for indx,epoch in enumerate(elmo_epochs):
    epoch_num = indx + 1

    cols = epoch.replace("\n", "").split("- ")
    cols = [i.split() for i in cols]

    elmo_train_losses.append(float(cols[2][1]))
    elmo_train_accuracies.append(float(cols[3][1]))
    elmo_val_losses.append(float(cols[4][1]))
    elmo_val_accuracies.append(float(cols[5][1]))
# Plotting the data
from matplotlib import pyplot as plt

x_points = list(range(1, 101))

plt.figure(figsize=(10, 10))
plt.plot(x_points, elmo_train_losses, label='Training Losses')
plt.plot(x_points, elmo_train_accuracies, label='Training Accuracies')
plt.plot(x_points, elmo_val_losses, label='Validation Losses')
plt.plot(x_points, elmo_val_accuracies, label='Validation Accuracies')
plt.title("ELMo")
plt.legend()
plt.show()
```

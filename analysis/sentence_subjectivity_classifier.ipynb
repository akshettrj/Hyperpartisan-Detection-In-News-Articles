```python
'''
Analysis of different Sentence Subjectivity Determining model
'''
```

```python
# Load all the data
all_epochs = []
with open("./stats/sentence_subjectivity_epochs.txt", "r") as data_file:
    all_epochs = data_file.readlines()
```

```python
# Gather data
train_losses = []
train_accuracies = []
val_losses = []
val_accuracies = []

for indx,epoch in enumerate(all_epochs):
    epoch_num = indx + 1

    cols = epoch.replace("\n", "").split(", ")
    cols = [i.split() for i in cols]

    train_losses.append(float(cols[1][1]))
    train_accuracies.append(float(cols[2][1]))
    val_losses.append(float(cols[3][1]))
    val_accuracies.append(float(cols[4][1]))
```

```python
# Plotting the data
from matplotlib import pyplot as plt

x_points = list(range(1, 501))

plt.figure(figsize=(10, 10))
plt.plot(x_points, train_losses, label='Training Losses')
plt.plot(x_points, train_accuracies, label='Training Accuracies')
plt.plot(x_points, val_losses, label='Validation Losses')
plt.plot(x_points, val_accuracies, label='Validation Accuracies')
plt.title("Subjectivity Classifier")
plt.legend()
plt.show()
```
